-How does the number of claims differ across locations?

![image-4](image-4.png)

There are 4 locations within the data set where claims where made. Sao Luis is the location with the highest number of claims(30), Recife with 25, Fortaleza with 22 and Natal has the lowest number of claims with 21. This gives a total of 98 claims against Vivendo.

-What is the distribution of time to close claims?
![image-10](image-10.png)

The histogram above shows that most of the cases were closed under 2400 days, with 2 outliers at 2997 and 3591 days. The most common ranges for cases to close where 200-400 days and 400-600 days with 16 cases each. This could possibly suggest that the legal team should aim to close future cases between 200-600 days.

-How does the average time to close claims differ by location?

![image-11](image-11.png)

The boxplots above display the distribution of the closing times per location with a line across them portraying the city's average.  Sao Luis has the highest average closing time at 1030 days and also has the highest IQR of 1405, which contributes to the fact Sao Luis doesn't have outliers. The other cities have 2-3 outliers with averages between 580-860 days. Their IQRs range between 440-590 suggesting Sao Luis has the most inconsistent data which in turn implies that the legal team should focus more on cases in Sao Luis. Further analysis should be carried using cases from Sao Luis in order to determine if there are significant patterns that are yet to be addressed.
